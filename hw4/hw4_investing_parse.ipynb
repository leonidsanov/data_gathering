```python
# Импорт библиотек
import requests
from lxml import html
```

```python
resp = requests.get(url='https://etfbaza.ru/list/', headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.5845.1024 YaBrowser/23.9.1.1024 (corp) Yowser/2.5 Safari/537.36'})
print(resp.status_code)
```

```python
tree = html.fromstring(html=resp.content)
```

```python
etf = tree.xpath("//tbody/tr")
```

```python
all_etf = []
for fund in etf:
  el = {'ticker': fund.xpath(".//td[@class='column-1']/descendant::text()"),
      'ISIN': fund.xpath(".//td[@class='column-2']/descendant::text()"),
      'title': fund.xpath(".//td[@class='column-3']/descendant::text()"),
      'class': fund.xpath(".//td[@class='column-4']/descendant::text()"),
      'provider': fund.xpath(".//td[@class='column-5']/descendant::text()"),
      'region': fund.xpath(".//td[@class='column-6']/descendant::text()"),
      'benchmark': fund.xpath(".//td[@class='column-7']/descendant::text()"),
      'commission': fund.xpath(".//td[@class='column-8']/descendant::text()"),
      'strategy': fund.xpath(".//td[@class='column-9']/descendant::text()"),
      'status': fund.xpath(".//td[@class='column-10']/descendant::text()"),
  }

  all_etf.append(el)

print(all_etf)
print(len(all_etf))
```

```python
from pymongo import MongoClient

def insert_to_db(list_etf):
    client = MongoClient('mongodb://localhost:27017/')
    db = client['funds']
    collection = db['ru_etf']
    collection.insert_many(list_etf)
    client.close
```

```python
insert_to_db(all_etf)
```
