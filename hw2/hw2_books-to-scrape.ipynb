```python
from bs4 import BeautifulSoup
import requests
import urllib.parse
import re
```

```python
name = [] # название книги
price = [] # цена книги
quantity = [] # количество книг в остатке
description = [] # аннотация
output = {} # словарь, в который добавляются все данные всех книг

url = 'http://books.toscrape.com/catalogue/page-1.html'

while True:

    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    result = soup.find_all('li', ('class', 'col-xs-6 col-sm-4 col-md-3 col-lg-3'))
    
    try:
        next_page_link = soup.find('li', ('class', 'next')).find('a')
    except AttributeError:
        break
    
    url_1 = 'http://books.toscrape.com/'
    url_2 = []
    for i in result:
        for link in i.find_all('div', ('class', 'image_container')):
            url_2.append(link.find('a').get('href'))

    url_joined = []

    for link in url_2:
        url_joined.append(urllib.parse.urljoin(url_1, link))

    for i in url_joined:
        response = requests.get(i)

        soup = BeautifulSoup(response.content, 'html.parser')

        # Парсинг названия товара. Обработка исключения: добавляем пустую строку.
        try:
            name.append(soup.find('h1').text)
        except:
            name.append('')

        # Парсинг цены товара.
        try:
            p = soup.find('p', ('class','price_color')).text
            p = float(re.sub(r'[^\d.]+', '', p))
            price.append(p)
        except:
            price.append('')

        # Парсинг количества товара в наличии.
        try:
            q = soup.find('p', ('class','instock availability')).get_text(strip=True)
            q = int(re.findall(r'\d+', q)[0])
            quantity.append(q)
        except:
            quantity.append('')

        # Парсинг описания товара.
        try:
            d = soup.find('div', id='product_description').find_next_sibling('p').text
            description.append(d)
        except:
            description.append('')

        # Записываем данные в словарь
        output = {'Name' : name, 'Price' : price, 'Quantity' : quantity, 'Description' : description}

url = url_1 + next_page_link['href']
```
