```python
# Импорт библиотек
import requests
from lxml import html
```

```python
resp = requests.get(url='https://en.wikipedia.org/wiki/Index_of_Economic_Freedom', headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.5845.1024 YaBrowser/23.9.1.1024 (corp) Yowser/2.5 Safari/537.36'})
print(resp.status_code)
```

```python
tree = html.fromstring(html=resp.content)
```

```python
countries = tree.xpath("//tbody/tr")
```

```python
all_countries = []
for country in countries:
  el = {'name': country.xpath(".//td[1]/a/text()"),
      'rank': country.xpath(".//td[2]/text()"),
      'score': country.xpath(".//td[3]/text()"),
  }

  all_countries.append(el)

print(all_countries)
print(len(all_countries))
```
